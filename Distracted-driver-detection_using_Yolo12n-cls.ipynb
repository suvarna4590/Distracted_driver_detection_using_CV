```python
from ultralytics import YOLO
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import os
from PIL import Image

# Load trained model
model = YOLO("/kaggle/working/runs/classify/train/weights/best.pt")

# Validation folder path
val_dir = "/kaggle/input/datasets/suvarnachiluka/distracted-driver-dataset/distracted-driver-split/val"

class_names = sorted(os.listdir(val_dir))

y_true = []
y_pred = []

for class_index, class_name in enumerate(class_names):
    class_folder = os.path.join(val_dir, class_name)

    for image_name in os.listdir(class_folder):
        image_path = os.path.join(class_folder, image_name)

        try:
            image = Image.open(image_path).convert("RGB")

            results = model.predict(image, imgsz=224, verbose=False)
            predicted_class = results[0].probs.top1

            y_true.append(class_index)
            y_pred.append(predicted_class)

        except Exception as e:
            print("Error processing:", image_path, e)

# Accuracy
accuracy = accuracy_score(y_true, y_pred)

print("Accuracy:", accuracy * 100)

# Precision, Recall, F1-score
print(classification_report(
    y_true,
    y_pred,
    target_names=class_names,
    digits=4
))

# Confusion Matrix
print(confusion_matrix(y_true, y_pred))
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average="weighted")
recall = recall_score(y_true, y_pred, average="weighted")
f1 = f1_score(y_true, y_pred, average="weighted")

print(f"Accuracy  : {accuracy * 100:.2f}%")
print(f"Precision : {precision * 100:.2f}%")
print(f"Recall    : {recall * 100:.2f}%")
print(f"F1-score  : {f1 * 100:.2f}%")
```
